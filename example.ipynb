hms11
===

hms11.py is a python-translated version of the simple stellar structure code:

http://www.astro.princeton.edu/~gk/A403/hms11.f

The original Fortran code is attributed to B. Paczynski, 1996.

Minor modifications have been made to adapt certain parts of it to Python, for clarity purposes.

```python
import hms11
```

```python
hms11.sch(0.71,0.02,1.,3.679,0.,7.3,2.1,outer_boundary_rho=1e-8,output_filename="test_solar")
```
